```python
grad_ret = pd.read_sql("SELECT COMP_5YRS, RETENTION FROM education15_table", database_connection)
grad_ret[grad_ret.RETENTION > 10]

# Remove the rows containing non-finite values
grad_ret = grad_ret[np.isfinite(grad_ret['COMP_5YRS'])]
grad_ret = grad_ret[np.isfinite(grad_ret['RETENTION'])]

ggplot(grad_ret, aes(x= 'RETENTION', y= 'COMP_5YRS')) +\
    geom_jitter(alpha=0.5)
```

```python
grad_ret['RETENTION'].corr(grad_ret['COMP_5YRS'])
```

We can see that retention has a correlation with completion rate.

```python
comp_rate = pd.read_sql("SELECT COMP_5YRS, REGION, OUT_ST_TUITION, FUNDING FROM education15_table", database_connection)

# Remove the rows containing non-finite values
comp_rate = comp_rate[np.isfinite(comp_rate['COMP_5YRS'])]
comp_rate = comp_rate[np.isfinite(comp_rate['OUT_ST_TUITION'])]

ggplot(comp_rate, aes(x= 'OUT_ST_TUITION', y= 'COMP_5YRS', colour = 'FUNDING')) +\
    geom_jitter(alpha=0.5)
```

We can see that for profit universities have the lowest completion rates and tuition.
Non profit universities have the highest completion rates and tuition while public universities fall somewhere between them.

```python
ggplot(comp_rate, aes(x= 'OUT_ST_TUITION', y= 'COMP_5YRS', colour= 'FUNDING')) +\
    geom_jitter(alpha=0.5) +\
    facet_wrap('REGION')
```

```python
com_ret_fund = pd.read_sql("SELECT COMP_5YRS, RETENTION, FUNDING FROM education15_table", database_connection)

# Remove the rows containing non-finite values
com_ret_fund = com_ret_fund[np.isfinite(com_ret_fund['COMP_5YRS'])]
com_ret_fund = com_ret_fund[np.isfinite(com_ret_fund['RETENTION'])]
com_ret_fund = com_ret_fund[np.isfinite(com_ret_fund['FUNDING'])]

ggplot(com_ret_fund, aes(x= 'RETENTION', y= 'COMP_5YRS', colour= 'FUNDING')) +\
       geom_jitter(alpha=0.5) +\
       facet_wrap('FUNDING') +\
       ggtitle('Completion Rate by Retention') 
```

We see that for public colleges, there is a directly proportional relationship between completion rate and retention.
For non-profit colleges, there is a linear correlation but it is not as clearly related as for public colleges.
For profit colleges don’t have any relation between completion rate and retention.
