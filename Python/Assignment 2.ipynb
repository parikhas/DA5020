```python
# Import the required libraries
import numpy as np
import pandas as pd
```

```python
# Load the file into a pandas data frame
df1 = pd.read_csv("surveys.csv")
df1.head()
```

```python
#1. Extract the survey observations for the first three months of 1990

winter_1990 = df1.loc[(df1['year'] == 1990) & df1['month'].isin([1, 2, 3])]
winter_1990.head()
```

```python
#2. Sort the 1990 winter surveys data by descending order of record ID, then by ascending order of weight

winter_1990.sort_values(['record_id', 'weight'], ascending=[False, True])
```

```python
#4. Write code that returns the average weight and hindfoot length of Dipodomys merriami (DM) individuals observed in each month

hw = df1.iloc[:,[1,5,7,8]]
DM_hw = hw.loc[(hw['species_id'] == 'DM')]
DM_hw.groupby(['month']).mean()
```
